```python
def plot_given_df_p(dataframe,lang1x,lang1y,lang2x,lang2y,lang3x,lang3y):
    # Separate values to make it easier to plot
    sep_values = pd.concat((pd.DataFrame(dataframe[col].tolist()) for col in dataframe), axis=1)
    #Now that the columns are created they need to be renamed
    sep_values.columns = [lang1x, lang1y, lang2x, lang2y, lang3x, lang3y]
    #convert to numeric for plotting
    sep_values[sep_values.columns[0:]] = sep_values[sep_values.columns[0:]].apply(pd.to_numeric, errors = 'coerce')
    #Print out plot of data
    c1=['#EA2027','#ED4C67']
    ax1 = sep_values.plot(y=[lang1x,lang1y],use_index=True, rot=0,color=c1,figsize=(10, 6), title='English Probability A')
    ax1.set_xlabel("Layers")
    ax1.set_ylabel("Classifier Accuracy")
    c2= ['#1e3799','#4a69bd']
    ax2 = sep_values.plot(y=[lang2x,lang2y],use_index=True, rot=0, color=c2,figsize=(10, 6), title='German Probability A')
    ax2.set_xlabel("Layers")
    ax2.set_ylabel("Accuracy")
    c3=['#079992','#38ada9']
    ax3 = sep_values.plot(y=[lang3x,lang3y],use_index=True, rot=0, color=c3,figsize=(10, 6), title='French Probability A')
    ax3.set_xlabel("Layers")
    ax3.set_ylabel("Classifier Accuracy")
    return ax1,ax2,ax3 

def plot_given_df_np(dataframe,lang1x,lang1y,lang2x,lang2y,lang3x,lang3y):
    # Separate values to make it easier to plot
    sep_values = pd.concat((pd.DataFrame(dataframe[col].tolist()) for col in dataframe), axis=1)
    #Now that the columns are created they need to be renamed
    sep_values.columns = [lang1x, lang1y, lang2x, lang2y, lang3x, lang3y]
    #convert to numeric for plotting
    sep_values[sep_values.columns[0:]] = sep_values[sep_values.columns[0:]].apply(pd.to_numeric, errors = 'coerce')
    #Print out plot of data
    c4=['#e58e26','#f6b93b']
    ax1 = sep_values.plot(y=[lang1x,lang1y],use_index=True, rot=0,color=c4,figsize=(10, 6), title='Korean Probability A')
    ax1.set_xlabel("Layers")
    ax1.set_ylabel("Classifier Accuracy")
    c5=['#57606f','#a4b0be']
    ax2 = sep_values.plot(y=[lang2x,lang2y],use_index=True, rot=0, color=c5,figsize=(10, 6), title='Portuguese Probability A')
    ax2.set_xlabel("Layers")
    ax2.set_ylabel("Accuracy")
    c6= ['#2ed573','#009432']
    ax3 = sep_values.plot(y=[lang3x,lang3y],use_index=True, rot=0, color=c6,figsize=(10, 6), title='Hebrew Probability A')
    ax3.set_xlabel("Layers")
    ax3.set_ylabel("Classifier Accuracy")
    return ax1,ax2,ax3 


plot_given_df_p(pro_group,"EngSA", "EngSO", "GerSA","GerSO","FreSA","FreSO")
plot_given_df_np(non_pro_group,"KorSA","KorSO","PorSA","PorSO","HebSA","HebSO")
```

```python
## Aggregating SO and SA into one for each language pair

#Flip the CSV

col_names = ["Bul","Cze","Slo","Cat","Ita","Rom"]
df = pd.read_csv('agg_bul.csv', names=col_names, encoding="ISO-8859-1")
df.drop(index=[0], inplace=True)
df = df.iloc[::-1]
display(df)
```

```python
def plot_given_df_agg(dataframe,lang1,lang2,lang3,lang4,lang5,lang6):
    # Separate values to make it easier to plot
    sep_values = pd.concat((pd.DataFrame(dataframe[col].tolist()) for col in dataframe), axis=1)
    #Now that the columns are created they need to be renamed
    sep_values.columns = [lang1,lang2,lang3,lang4,lang5,lang6]
    #convert to numeric for plotting
    sep_values[sep_values.columns[0:]] = sep_values[sep_values.columns[0:]].apply(pd.to_numeric, errors = 'coerce')
    #Print out plot of data
    c_list=['#EA2027', '#1e3799', '#079992', '#e58e26','#57606f', '#2ed573']
    ax1 = sep_values.plot(y=[lang1,lang2,lang3,lang4,lang5,lang6],use_index=True, rot=0,color=c_list,figsize=(10, 6), title='Probability A Averages on Romanian Classifier')
    ax1.set_xlabel("Layers")
    ax1.set_ylabel("Probability of A given S Averages")
    return ax1

plot_given_df_agg(df,"Bul","Cze","Slo","Cat","Ita","Rom")
```
